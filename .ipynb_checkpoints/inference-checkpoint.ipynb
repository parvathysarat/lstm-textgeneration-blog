```python
import argparse
import codecs
import chardet
import numpy as np
from text_generation import Lyrics_model,Lyrics_preprocess,Lyrics_Batch
from collections import deque
import tensorflow as tf

def main():
#     parser=argparse.ArgumentParser()
#     parser.add_argument('--model_name', type=str, required=True)
#     parser.add_argument('--vocabulary_file', type=str, required=True)
#     parser.add_argument('--output_file', type=str, required=True)
    
#     parser.add_argument('--seed', type=str, default="Yea")
#     parser.add_argument('--sample_length', type=int, default=1500)
#     parser.add_argument('--log_frequency', type=int, default=100)
#     args=parser.parse_args()
    
#     model_name=args.model_name
#     vocabulary_file=args.vocabulary_file
#     output_file=args.output_file
#     seed=args.seed.decode('utf-8')
#     sample_length=args.sample_length
#     log_frequency=args.log_frequency
    
    
    model=Lyrics_model.Model("My_Model")
    model.restore()
    classifier=model.get_classifier()
    
    
    vocabulary=Lyrics_preprocess.Preprocessing()
    vocabulary.retrieve("blogdata_nonextline.vocab")
    seed="medical college"
    sample_length=1500
    log_frequency=100
    output_file="Output.txt"
    
    sample_file=codecs.open(output_file, 'w', 'utf-8')
    
    stack=deque([])
    for i in range(0, model.sequence_length-len(seed)):
        stack.append(u' ')
    print ("Vocabulary - space - ",list(vocabulary.vocabulary)[1])
    for char in seed:
#         char="b'"+char+"'"
        if char=="b' '":
            print("found",char)
        if ("b'"+char+"'") not in vocabulary.vocabulary:
            print (char,"is not in vocabulary file")
            char=u' '
        stack.append(char)
        sample_file.write(char)
      
    with tf.Session() as sess:
        tf.global_variables_initializer().run()
        
        saver=tf.train.Saver(tf.global_variables())
        ckpt=tf.train.get_checkpoint_state("My_Model")
        
        if ckpt and ckpt.model_checkpoint_path:
            saver.restore(sess,ckpt.model_checkpoint_path)
            
            
            for i in range(0, sample_length):
                
                vector=[]
#                 print (stack)
#                 stack=stack.strip()
                for char in stack:
                    print(char,type(char))
#                     encoding = chardet.detect(char)
#                     print(encoding['encoding'],char)
#                     print(char.encode('utf-8'))
                    if ' ' in char or '.' in char or '\n'in char:
                        if '.' in char:
                            vector.append(vocabulary.binary_vocabulary["b'\u002E'"])
                        elif ' ' in char:    
                            vector.append(vocabulary.binary_vocabulary["b'\u0020'"])
                        elif '\n' in char:
                            vector.append(vocabulary.binary_vocabulary["b'\u005Cn'"])
                    else:
                        if char=='w' or char=='e' or char=='l':
                            
                            vector.append(vocabulary.binary_vocabulary[str(char.encode('utf-8'))])
                            print(char,"YES")
                            
                            
                        else:
                            vector.append(vocabulary.binary_vocabulary[str(char.encode('utf-8'))])
                            print(char,"CONVERTED")
                            
#                         char="b'"+char+"'"
                        
#                         print(str(char.decode('utf-8')))
#                         
#                         vector.append(vocabulary.binary_vocabulary[str(char.encode('utf-8'))])
                vector=np.array([vector])
                prediction=sess.run(classifier, feed_dict={model.x:vector})
                predicted_char=vocabulary.char_lookup[np.argmax(prediction)]
                
                
                stack.popleft()
                lis=[ch for ch in seed]
                if predicted_char in lis:
                    print("pred",predicted_char,len(predicted_char))
                    stack.append(predicted_char)
                else:
                    stack.append(predicted_char[2:3])
                
                sample_file.write(predicted_char[2:3])
                
                
                if i%log_frequency==0:
                    print ("Progress: {}%". format((i*100)/sample_length))
                    
            sample_file.close()
            print ("Lyrics saved in {}".format(output_file))
        
        
if __name__=="__main__":
    main()
```
